```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt


#Get the data
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

# Explore the data to gain insights
# Plotting scatter plot of BMI vs. Disease Progression
plt.scatter(diabetes_X[:, 2], diabetes_y)
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Scatter plot of BMI vs. Disease Progression')
plt.show()

# Fine-tune your models and combine them into a great solution
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(diabetes_X[:, 2], diabetes_y, test_size=0.2, random_state=42)

# Train the linear regression model
regressor = LinearRegression()
regressor.fit(X_train.reshape(-1, 1), y_train.reshape(-1, 1))

# Print the attribute names
attribute_names = datasets.load_diabetes().feature_names
print("Attribute Names:")
for name in attribute_names:
    print(name)

# Print the progression values
print("\nDisease Progression:")
for progression in diabetes_y:
    print(progression)

# Present your solution
print('Model Parameters:')
print('Slope:', regressor.coef_[0][0])
print('Intercept:', regressor.intercept_[0])

# Calculate and print the loss on the training and test datasets
train_loss = mean_squared_error(y_train, regressor.predict(X_train.reshape(-1, 1)))
test_loss = mean_squared_error(y_test, regressor.predict(X_test.reshape(-1, 1)))

print('Train Loss:', train_loss)
print('Test Loss:', test_loss)

train_r2 = regressor.score(X_train.reshape(-1, 1), y_train)
test_r2 = regressor.score(X_test.reshape(-1, 1), y_test)
print('Train R-squared:', train_r2)
print('Test R-squared:', test_r2)

# Plotting the training data, test data, and the linear regression model
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Test Data')
plt.plot(X_train, regressor.predict(X_train.reshape(-1, 1)), color='red', linewidth=2, label='Linear Regression')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.legend()
plt.title('Linear Regression: BMI vs. Disease Progression')
plt.show()
```
