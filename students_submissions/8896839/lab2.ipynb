```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

**Matplotlib** : *Plotting Cosine function*


```python
X = np.linspace(0, 2*np.pi, 100)
Y = np.cos(X)

fig, ax = plt.subplots()
ax.plot(X,Y, color = 'blue')
```

**Matplotlib** : *Generate random data for the x-axis (time) and y-axis (temperature)*

```python
# Generate random data for the x-axis (time) and y-axis (temperature)
time = list(range(10))
temperature = np.random.randint(-3,30, 10)

# Create the line graph
plt.plot(time, temperature, marker='o')

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Variation')

# Display the graph
plt.show()
```

**Plotly** : *Bubble charts, color and size data are added to hover information*

Loading iris dataset and plotting using scatter plot by plotly

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size='petal_length', hover_data=['petal_width'])
fig.show()
```

Reference : https://plotly.com/python/line-and-scatter/

**Seaborn** 

```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```
