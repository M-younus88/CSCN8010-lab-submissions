```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import cross_val_score
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
import numpy as np
```

#### Loading the datasets

```python
# load diabetes dataset
Xdata, ydata = load_diabetes(as_frame=True, scaled=False, return_X_y=True)
Xdata
```

### 1. Utilize the diabetes dataset from lab 4. Perform cross-validation on nine polynomial models, ranging from degree 0 to 8.

```python
degrees = range(9)
results = []

for degree in degrees:
    poly_features = PolynomialFeatures(degree=degree)
    X_poly = poly_features.fit_transform(Xdata)

    model = LinearRegression()
    scores = cross_val_score(model, X_poly, ydata, cv=5, scoring='r2')
    mean_r2 = np.mean(scores)
    std_r2 = np.std(scores)

    scores_mae = cross_val_score(
        model, X_poly, ydata, cv=5, scoring='neg_mean_absolute_error')
    mean_mae = -np.mean(scores_mae)
    std_mae = np.std(scores_mae)

    results.append((degree, mean_r2, std_r2, mean_mae, std_mae))
```

### 2. Construct a table summarizing the cross-validation results. Each model should have a separate row in the table. Include the R-Squared and Mean Absolute Error (MAE) metrics for each model. Calculate the mean value and standard deviation of these metrics from the cross-validation. Include both values.

```python

#cross validation results
results_table = pd.DataFrame(results, columns=[
                             'Degree', 'R-Squared (Mean)', 'R-Squared (Std)', 'MAE (Mean)', 'MAE (Std)'])

results_table
```

```python
# Calculate the mean and standard deviation of the metrics

mean_r2_all = results_table['R-Squared (Mean)'].mean()
std_r2_all = results_table['R-Squared (Mean)'].std()
mean_mae_all = results_table['MAE (Mean)'].mean()
std_mae_all = results_table['MAE (Mean)'].std()


summary = pd.DataFrame(
    {'Metric': [ 'R-Squared', 'MAE'],
     'Mean': [mean_r2_all, mean_mae_all],
     'Standard Deviation': [ std_r2_all, std_mae_all]
     }
)

summary
```

### 3. Identification of the Best Model: Identify the model that exhibits the highest performance based on the R-Squared and MAE metrics. Provide an explanation for choosing this specific model.

```python
best_model_r2 = results_table.loc[results_table['R-Squared (Mean)'].idxmax()]
best_model_mae = results_table.loc[results_table['MAE (Mean)'].idxmin()]

print(f"\nBest model based on R-Squared: {best_model_r2}")
print(f"Best model based on MAE: {best_model_mae}")
```
